\begin{aligned}
x_1 \ldots x_n|\mu &\sim N(\mu, 1) 
\\
\mu &\propto 1
\\
\mu|x_1 \ldots x_n &\sim N(\bar{x}, 1/n)
\end{aligned}

```r
set.seed(1)
source("normal-normal.R")

n <- 100
S <- 5000
x <- rnorm(n, 0, 1)

nn_samples <- hmc(param_init = 3, 
                  U = nn_U(x), 
                  dU = nn_dU(x), 
                  e = 0.001, 
                  L = 50, 
                  iters = S)
```

```r
hist(nn_samples$samples[101:S,], probability = TRUE, 
     main = "HMC vs Analytical Posterior", xlab = "x", breaks = 25)
plot_grid <- seq(-.5, .5, .01)
lines(plot_grid, dnorm(plot_grid, mean(x), sqrt(1 / n)))
```
